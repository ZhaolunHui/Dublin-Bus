```python
#Extract trip id's from trips table for bus line 46A and for the month of November and December
def extract_tripsid_46A():
    sql = "select distinct trips100.TRIPID,trips100.LINEID from mydbservlet.trips100 where trips100.LINEID='46A' and DAYOFSERVICE like '%NOV%'" 
    cursor = db.cursor()
    cursor.execute(sql)
    result = cursor.fetchall()
    return result
```

#### Use the datetime to get corresponding day of the week

```python
date = df_46A[[0]]
```

```python
date = list(date[0])
```

```python
date[0]
```

```python
extract_date = []
for i in date:
    new_date = i[0:9]
    extract_date.append(new_date)
    
```

```python
import datetime
extract_week_day = []
for i in extract_date:
    extract_week_day.append(datetime.datetime.strptime(i, '%d-%b-%y').strftime('%A'))
```

```python
#Append a new column named 'Extracted_Date'
df_46A = df_46A.assign(Extracted_Date=extract_date)
```

```python
#Append a new column named 'Day' indicating day of the week
df_46A = df_46A.assign(Day=extract_week_day)
```

```python
#Renaming the columns
df_46A.rename({0:'DateTime', 2: 'Sequence',3:'StopID',4:'Planned_Arr',5:'Actual_Arr'},axis=1,inplace=True) 
```

```python
df_46A.head()
```

```python
#Creating Dummies for the day of the week
data_input1= pd.DataFrame(df_46A['Day'])
```

```python
dummy=pd.get_dummies(data_input1)
```

```python
df_46A=pd.concat([df_46A,dummy],axis=1)
```

```python
df_46A.head(5)
```

```python
df_46A.dtypes
```

```python
continuous_features = df_46A[['Sequence','StopID','Planned_Arr']].columns
# Convert data type to category for these columns
for column in continuous_features:
    df_46A[column] = df_46A[column].astype('int64')  
```

```python
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
pd.options.mode.chained_assignment=None
%matplotlib inline
```

```python
input_to_model=pd.DataFrame(df_46A[['Sequence','StopID','Planned_Arr','Day_Monday','Day_Thursday','Day_Tuesday','Day_Wednesday']])
#Day_Friday has been taken out because if mon-sun all are zeroes that means friday is 1. Using dummy one hot encoding concept
output = df_46A['Actual_Arr']
X_train,X_test,Y_train,Y_test=train_test_split(input_to_model,output,test_size=0.33,random_state=42)
print (X_train.shape)
```

#### Fine tuning the model using GridSearchCV and Parameter Grid 

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
{'n_estimators': [3, 10, 20], 'max_features': [2, 4, 6],'max_depth': [4,8,12]},

{'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4],'max_depth': [4,8]},
]
forest_reg = RandomForestRegressor(random_state=42)
# train across 5 folds, that's a total of (12+6)*5=90 rounds of training 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                       scoring='neg_mean_squared_error')
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_params_
```

#### Train the dataset using Random Forest Regresor based on the parameters obtained above

```python

rf=RandomForestRegressor(n_estimators=20,max_depth=12,max_features=6)
rf.fit(X_train,Y_train)
print(Y_test.shape)
```

```python
prediction=rf.predict(X_test)
```

#### Model Evaluation Parameters: RMSE and MSLE

```python
print(np.sqrt(metrics.mean_squared_error(Y_test,prediction)))
```

- The model shows a root mean squared error of approximately 345 seconds or 5 mins 45 seconds

```python
from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(Y_test,prediction)
```

####  Evaluating the model using k-fold cross-validation on the full dataset

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

#cross_val_score is an inbuilt function that calculates the accuracy score of each fold
#Variable all_random_scores is to store the list of accuracy scores of each fold
#Parameter cv=10 implies 10 fold cross validation
all_random_scores = cross_val_score(RandomForestRegressor(n_estimators=20,max_depth=12,max_features=6),input_to_model,output,cv=10)
```

```python
for i in range(len(all_random_scores )):
    print("Score of ",i+1," fold ",all_random_scores[i]*100,"%")
```

```python
mean_accuracy_random_regression = np.mean(all_random_scores)

print("Mean Score of the Random Forest Regressor Model after applying 10 folds cross validation is " ,mean_accuracy_random_regression *100,"%")
```

```python
#cross_val_predict is an in-built function that does out-of-fold predictions for the complete dataset
#cross_val_predict() returns, for each element in the input, the prediction that was obtained for that element when it was in the test set.
#We use cv=10 for 10 fold cross validation
y_pred_random_df = cross_val_predict(RandomForestRegressor(n_estimators=20,max_depth=12,max_features=6),input_to_model,output,cv=10)
```
